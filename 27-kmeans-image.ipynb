```python
# 使用K-means对图像进行聚类，并显示聚类压缩后的图像
import numpy as np
import PIL.Image as image
from sklearn.cluster import KMeans
from sklearn import preprocessing
import matplotlib.image as mping
```

### 加载数据

```python
f= open('./data/kmeans-master/weixin.jpg','rb')
data =[]
img = image.open(f)

# 得到图像尺寸
width ,height = img.size
# 得到点(x,y)的三个通道值
for x in range(width):
    for y in range(height):
        c1,c2,c3 = img.getpixel((x,y))
        data.append([c1,c2,c3])
        
f.close()
```

```python
img.size
```

```python
data
```

```python
mm = preprocessing.MinMaxScaler()
```

```python
data_mm = mm.fit_transform(data)
data_mm
```

### 模型

```python
data_mat = np.mat(data_mm)
data_mat
```

```python
# 用K-Means对图像进行2聚类
kmeans2 = KMeans(n_clusters=2)
kmeans2.fit(data_mat)
label2 = kmeans2.predict(data_mat)
```

```python
label2
```

```python
# 将图像聚类结果，转化成图像尺寸的矩阵
label2 = label2.reshape([width,height])
label2
```

```python
# 创建个新图像pic_mark，用来保存图像聚类的结果，并设置不同的灰度值
pic_kmeans2 = image.new("L",(width,height))
for x in range(width):
    for y in range(height):
        pic_kmeans2.putpixel((x,y),int(256/(label2[x][y]+1))-1)
pic_kmeans2.save('./data/kmeans-master/pic_kmeans2.jpg','JPEG')
```

```python
# 用K-Means对图像进行16聚类
kmeans16 = KMeans(n_clusters=16)
kmeans16.fit(data_mat)
label16 = kmeans16.predict(data_mat)
label16
```

```python
# 将图像聚类结果，转化成图像尺寸的矩阵
label16 = label16.reshape([width,height])
label16
```

```python
from skimage import color
```

```python
# 将聚类标识矩阵转化为不同颜色的矩阵a
label_color = (color.label2rgb(label16)*255).astype(np.uint8)
label_color = label_color.transpose(1,0,2)
images = image.fromarray(label_color)
images.save('./data/kmeans-master/pic_kmeans16.jpg','JPEG')
```

```python
kmeans16.cluster_centers_
```

```python
# 创建个新图像img，用来保存图像聚类压缩后的结果
img_kemans16= image.new("RGB",(width,height))
for x in range(width):
    for y in range(height):
        c1 = kmeans16.cluster_centers_[label16[x, y], 0]
        c2 = kmeans16.cluster_centers_[label16[x, y], 1]
        c3 = kmeans16.cluster_centers_[label16[x, y], 2]
        img_kemans16.putpixel((x,y), (int(c1*256)-1, int(c2*256)-1, int(c3*256)-1))
img_kemans16.save("./data/kmeans-master/img_kemans16.jpg")
```
